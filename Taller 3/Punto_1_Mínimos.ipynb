```python
import numpy
import sympy 
import matplotlib.pyplot 
from mpl_toolkits.mplot3d import axes3d as ax
```

```python
def f1(x):
    return 2*x - 2

def f2(x):
    return (1 - x) / 2

def f3(x):
    return 4 - x


x0 = numpy.linspace(0, 19, 20)
y1 = f1(x0)
y2 = f2(x0)
y3 = f3(x0)


A = numpy.array([[2, -1], [1, 2], [1, 1]])
b = numpy.array([2, 1, 4])


AT = A.transpose()
prod = numpy.dot(AT, A)
inv_prod = numpy.linalg.inv(prod)
pt = numpy.dot(numpy.dot(inv_prod, AT), b)
print(pt)
```

```python
matplotlib.pyplot.plot(x0,y1)
matplotlib.pyplot.plot(x0,y2)
matplotlib.pyplot.plot(x0,y3)
matplotlib.pyplot.scatter(pt[0],pt[1],color='red')
```

```python
def f(x, y):
    a1 = 2*x - 2 - y
    a2 = x + 2*y - 1
    a3 = x + y - 4
    return numpy.sqrt(a1**2 + a2**2 + a3**2)

f = numpy.vectorize(f)

X = numpy.arange(-5, 5, 0.01)
Y = numpy.arange(-5, 5, 0.01)
X, Y = numpy.meshgrid(X, Y)
Z = f(X, Y)


fig = matplotlib.pyplot.figure()
ax = fig.add_subplot(111, projection='3d')


surf = ax.plot_surface(X, Y, Z)
ax.view_init(15, 20)


ax.scatter(pt[0], pt[1], f(pt[0], pt[1]), c='b')
```
