```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def E(x,y,z,a=0.5):
  
  funcionx = lambda phi: (x-a*np.cos(phi)) / ((x**2 + y**2 + z**2 + a**2)  - 2*a*x*np.cos(phi) - 2*a*y*np.sin(phi))**(3/2)
  funciony = lambda phi: (y-a*np.sin(phi)) / ((x**2 + y**2 + z**2 + a**2)  - 2*a*x*np.cos(phi) - 2*a*y*np.sin(phi))**(3/2)
  funcionz = lambda phi: (z) / ((x**2 + y**2 + z**2 + a**2)  - 2*a*x*np.cos(phi) - 2*a*y*np.sin(phi))**(3/2)

  c = 0
  b = 2*np.pi
  n = 30 
  r,p=np.polynomial.legendre.leggauss(n)

  variable = a*((b-c)*r+c+b)

  Ex= (1/(2*np.pi))*0.5*(b-c)*np.sum(p*funcionx(variable))
  Ey= (1/(2*np.pi))*0.5*(b-c)*np.sum(p*funciony(variable))
  Ez= (1/(2*np.pi))*0.5*(b-c)*np.sum(p*funcionz(variable))
  return[Ex,Ey,Ez]
E(0.6,0.6,1)

```

```python
N = 6

x=np.linspace(-0.6,0.6,N)
y=np.linspace(-0.6,0.6,N)
z=np.linspace(-0.6,0.6,N)

a = np.zeros((N,N,N))
b = np.zeros((N,N,N))
c = np.zeros((N,N,N))

for i in range(6):
  for j in range(6):
    for k in range(6):
      a[i][j][k] = E(x[i],y[j],z[k])[0]
      b[i][j][k] = E(x[i],y[j],z[k])[1]
      c[i][j][k] = E(x[i],y[j],z[k])[2]

print([a,b,c])
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')

for i in range(6):
    for j in range(6):
      for k in range(6):
        ax.quiver(x[i],y[j],z[k],a[i,j,k],b[i,j,k],c[i,j,k])

```
