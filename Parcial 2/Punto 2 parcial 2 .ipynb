```python
%matplotlib notebook
from mpl_toolkits.mplot3d import axes3d
from scipy import integrate
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
def CreateSphere(N,R=1):
    
    Points = np.zeros((N,3))
        
    for i in tqdm(range(N)):
        
        phi = np.random.uniform(0,2*np.pi)
        u = np.random.rand()
        r = R*u**(1/3)
        costheta = np.random.uniform(-1,1)
        theta = np.arccos(costheta)
        
        Points[i] = [r*np.cos(phi)*np.sin(theta),r*np.sin(phi)*np.sin(theta),r*np.cos(theta)]
        
    return Points
```

```python
N = 100000
Sphere = CreateSphere(N)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(Sphere[:,0],Sphere[:,1],Sphere[:,2])
```

```python
def f2(x,y,z):
    return np.exp( np.sqrt(x**2+y**2+z**2) )
X = Sphere[:,0]
Y = Sphere[:,1]
Z = Sphere[:,2]
```

```python
Ixx = np.sum(Y**2 + Z**2) / N
Iyy = np.sum(X**2 + Z**2) / N
Izz = np.sum(Y**2 + X**2) / N
```
