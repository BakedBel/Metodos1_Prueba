```python
n = 25
R = 2.0
x = np.linspace(-4,4,n)
y = np.linspace(-4,4,n)
X,Y = np.meshgrid(x,y)

def funcion(x,y):
    R = 2.0
    d = x**2 + y**2
    return 2*x*(1-(R**2)/2)
def derivate(f,x,y,h=0.001):
    solution = np.zeros(2)
    dx = (f(x+h,y)-f(x-h,y))/(2*h)
    dy = (f(x,y+h)-f(x,y-h))/(2*h)
    solution[0]= dx
    solution[1]=dy
    return solution
```
