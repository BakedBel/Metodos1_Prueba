```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

p1 = np.linspace(0.1, 0.9, 50)
p2 = np.linspace(0.1, 0.5, 25)
P1, P2 = np.meshgrid(p1, p2)

pcarasello = np.zeros(len(p1))

def probabilidad(p1, p2):
    pcarasello = 0.25*p1*(1-p2)+1*p2*(1-p1)
    return pcarasello

P = probabilidad(P1, P2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(P1, P2, P, cmap='coolwarm')
ax.set_xlabel('Probabilidad p1')
ax.set_ylabel('Probabilidad p2')
ax.set_zlabel('Probabilidad')
ax.set_title('Superficie de probabilidad del evento A')


plt.show()
```
