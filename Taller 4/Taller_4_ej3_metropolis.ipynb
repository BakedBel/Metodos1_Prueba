```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import scipy.integrate as integrate
import matplotlib.animation as anim
from statistics import mode
```

```python
def normal(x, mu, sigma):
    c = 1 / np.sqrt(2 * np.pi * sigma ** 2)
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2) * c
```

```python
def Metropolis(x0, Normal, NSteps=int(1e4), delta=0.4):
    x = np.zeros((NSteps, 1))
    x[0] = x0
    for i in tqdm(range(1, NSteps)):
        P0 = Normal(x[i - 1])
        xf = x[i - 1] + delta * 2 * (np.random.rand() - 0.5)
        P1 = Normal(xf)
        alpha = np.minimum(1, P1 / P0)
        g = np.random.rand()
        if alpha > g:
            x[i, 0] = xf
        else:
            x[i, :] = x[i - 1, :]

    return x[100:, :]
def Normal(x, mu=2, sigma=0.5):
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2) / (sigma * np.sqrt(2 * np.pi))
def run_metropolis(x0, mu=2, sigma=0.5, NSteps=int(1e4), delta=0.4):
    return Metropolis(x0, lambda x: Normal(x, mu, sigma), NSteps, delta)
```
